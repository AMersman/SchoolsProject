```python
#import the 'Building Level Per Pupil Expenditures 2019' data 
spend2019 = pd.read_excel('../raw_data/Building Level Per Pupil Expenditures 2019.xlsx')
```

```python
spend2019.head()
```

```python
#review the spend2019 DataFrame
spend2019.info()
```

```python
#use dropna(how='all') to drop empty columns
spend2019 = spend2019.dropna(axis='columns', how='all')
```

```python
spend2019[spend2019['District Name'].isnull()]
```

```python
#048921 is a charter school that is closed. We can omit this data.
spend2019 = spend2019.drop([1181, 1182])
spend2019.info()
```

```python
#omit columns from district only dataframe that relate to specific buildings
spend2019dist = spend2019[spend2019['Building Name']=='DISTRICT TOTALS']
spend2019dist=spend2019dist.iloc[: ,0:9]
spend2019dist.info()
```

```python
spend2019dist.head()
```

```python
spend2019dist = spend2019dist.drop(['Building Name', 'Building Number'], axis=1)
```

```python
spend2019dist.info()
```

```python
#import the 'Building Level Per Pupil Expenditures 2020' data 
spend2020 = pd.read_excel('../raw_data/Building Level Per Pupil Expenditures 2020.xlsx')
spend2020.info()
```

```python
spend2020 = spend2020.dropna(axis='columns', how='all')
spend2020.info()
```

```python
#omit columns from district only dataframe that relate to specific buildings
spend2020dist = spend2020[spend2020['Building Name']=='DISTRICT TOTALS']
spend2020dist=spend2020dist.iloc[: ,0:9]
spend2020dist.info()
```

```python
spend2020dist = spend2020dist.drop(['Building Name', 'Building Number'], axis=1)
spend2020dist.info()
```

```python
#the data for 2020 has 2 more rows, what rows are unique to spend2020dist
dif_list = [x for x in list(spend2020['District Name'].unique()) if x not in list(spend2019['District Name'].unique())]
```
