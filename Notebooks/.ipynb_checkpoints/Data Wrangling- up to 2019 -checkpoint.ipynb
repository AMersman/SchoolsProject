```python
#import school district data for MO
Dist = pd.read_csv('../raw_data/School District List.csv')
```

```python
#Rename columns and keep only code, name, city, and county
Dist = Dist.rename(columns={'DNAME':'District Name', 'DCITY':'City', 'DCOUNTY':'County', 'CTYDIST':'District Code'})
Dist = Dist[['District Code', 'District Name', 'City', 'County']]
Dist.columns
```

```python
#import 'District Accreditation' which has the district accreditation for each district in 2021
Accred = pd.read_excel('../raw_data/District Accreditation.xlsx')
```

```python
Accred.head()
```

```python
#Rename columns of Accred to match Dist and merge the dataframes
Accred = Accred.rename(columns = {'COUNTY_DISTRICT_CODE':'District Code', 'DISTRICT_NAME':'District Name', 'CLASSIFICATION':'Accreditation Status'})
Accred = Accred.drop(columns='SCHOOL_YEAR')
```

```python
#Change District Name for Nixa Public Schools to match formatting
Accred['District Name'] = Accred['District Name'].replace(['Nixa Public Schools'],'NIXA PUBLIC SCHOOLS')
Accred[Accred['District Name'].str.contains("^NIXA")]
```

```python
#Correct LONEDELL R-XIV from R-14
Accred['District Name'] = Accred['District Name'].replace(['LONEDELL R-14'],'LONEDELL R-XIV')
Accred[Accred['District Name'].str.contains("^LONE")]
```

```python
DistrictInfo = Accred.merge(Dist, how='left', on=['District Code'])
DistrictInfo.info()
```

```python
#review the DataFrame
DistrictInfo.head()
```

```python
DistrictInfo[DistrictInfo['Accreditation Status']!= 'A']
```

```python
#Formatting changes, drop District Name_y and rename District Name_x, change Accreditation Status Values
DistrictInfo = DistrictInfo.drop(columns='District Name_y')
DistrictInfo = DistrictInfo.rename(columns={'District Name_x':'District Name'})
DistrictInfo.loc[DistrictInfo["Accreditation Status"] == "A", "Accreditation Status"] = 'Accredited'
DistrictInfo.loc[DistrictInfo["Accreditation Status"] == "P", "Accreditation Status"] = 'Provisional'
```

```python
#Import School District size data
sqmiles = pd.read_excel('../raw_data/School District Area in Square Miles.xls', header=1)
sqmiles.info()
```

```python
sqmiles = sqmiles.rename(columns={'County District':'District Code', 'District':'District Name'})
sqmiles.head()
```

```python
#match formatting of other DF
sqmiles['District Name'] = sqmiles['District Name'].str.upper()
sqmiles['District Code'] = sqmiles['District Code'].astype('str')
sqmiles['District Code'] = sqmiles['District Code'].str.replace('-', '', regex=False)
sqmiles['District Code'] = sqmiles['District Code'].astype('int')
sqmiles['Area (Sq Miles)'] = sqmiles['Area (Sq Miles)'].round(2)
sqmiles.head()
```

```python
Districts = Districts.merge(sqmiles, how='left', on=['District Code', 'District Name'])
Districts.info()
```

```python
Districts.sort_values(by='Area (Sq Miles)')
```

```python
#fill in missing Area data
Districts.loc[Districts["District Name"] == "NEOSHO SCHOOL DISTRICT", "Area (Sq Miles)"] = 221.89
Districts.loc[Districts["District Name"] == "NORMANDY SCHOOLS COLLABORATIVE", "Area (Sq Miles)"] = 10.6
Districts.loc[Districts["District Name"] == "SPECL. SCH. DST. ST. LOUIS CO.", "Area (Sq Miles)"] = 510
Districts.loc[Districts["District Name"] == "WRIGHT CITY R-II OF WARREN CO.", "Area (Sq Miles)"] = 91.02
Districts.sort_values(by='Area (Sq Miles)')
```
