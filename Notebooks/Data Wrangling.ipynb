```python
import pandas as pd
```

```python
#import school district data for MO
Dist = pd.read_csv('../raw_data/School District List.csv')
```

```python
Dist.head()
```

```python
#Rename columns and keep only code, name, city, and county
Dist = Dist.rename(columns={'DNAME':'District Name', 'DCITY':'City', 'DCOUNTY':'County', 'CTYDIST':'District Code'})
Dist = Dist[['District Code', 'District Name', 'City', 'County']]
Dist.columns
```

```python
Dist.info()
```

```python
Dist[Dist['County'].isnull()]
```

```python
#import 'District Accreditation' which has the district accreditation for each district in 2021
Accred = pd.read_excel('../raw_data/District Accreditation.xlsx')
```

```python
Accred.head()
```

```python
#Rename columns of Accred to match Dist and merge the dataframes
Accred = Accred.rename(columns = {'COUNTY_DISTRICT_CODE':'District Code', 'DISTRICT_NAME':'District Name', 'CLASSIFICATION':'Accreditation Status'})
Accred = Accred.drop(columns='SCHOOL_YEAR')
```

```python
DistrictInfo = Accred.merge(Dist, how='left', on=['District Code'])
DistrictInfo.info()
```

```python
#review the DataFrame
DistrictInfo.head()
```

```python
DistrictInfo[DistrictInfo['Accreditation Status']!= 'A']
```

```python
#Formatting changes, drop District Name_y and rename District Name_x, change Accreditation Status Values
DistrictInfo = DistrictInfo.drop(columns='District Name_y')
DistrictInfo = DistrictInfo.rename(columns={'District Name_x':'District Name'})
DistrictInfo.loc[DistrictInfo["Accreditation Status"] == "A", "Accreditation Status"] = 'Accredited'
DistrictInfo.loc[DistrictInfo["Accreditation Status"] == "P", "Accreditation Status"] = 'Provisional'
```
